# 로컬

```python
import torch
from PIL import Image
from yolov5 import YOLOv5

model = YOLOv5("yolov5s.pt")  # YOLOv5 Small 모델 불러오기
img = Image.open("테스트 이미지 경로")  # 입력 이미지
results = model.predict(img)   # 객체 탐지 수행
results.show()                 # 결과 출력
```

# pytorch hub

```python
import torch
from PIL import Image

# 모델 불러오기 (예: 'yolov5s' 모델)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # PyTorch Hub를 통해 모델 불러오기

# 이미지 불러오기
img = Image.open('C:/Users/ailab/pytorch/study/image/냥이.jpg')

# 예측 수행
results = model(img)

# 예측 결과 출력
results.show()
```
